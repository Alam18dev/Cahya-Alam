```python
from google.colab import files

uploaded = files.upload()  # 5_6217725834027865017.csv

```

```python
import pandas as pd

file_name = list(uploaded.keys())[0]  # Ambil nama file yang diunggah
df = pd.read_csv(file_name)  # Baca file CSV
print(df.head())  # Cek beberapa data pertama
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set_style("whitegrid")

# Figure setup
fig, axes = plt.subplots(3, 2, figsize=(18, 15))
fig.delaxes(axes[2, 1])  # Hapus subplot yang kosong

# 1. Distribusi IMDb Score (Biru)
sns.histplot(df["imdb_score"], bins=20, kde=True, ax=axes[0, 0], color="blue")
axes[0, 0].set_title("Distribusi IMDb Score")
axes[0, 0].set_xlabel("IMDb Score")
axes[0, 0].set_ylabel("Jumlah Film")

# 2. Budget vs Gross Revenue (Merah)
sns.scatterplot(data=df, x="budget", y="gross", alpha=0.5, ax=axes[0, 1], color="red")
axes[0, 1].set_title("Budget vs Gross Revenue")
axes[0, 1].set_xlabel("Budget (USD)")
axes[0, 1].set_ylabel("Gross Revenue (USD)")

# 3. Likes Pemeran Utama vs IMDb Score (Hijau)
sns.scatterplot(data=df, x="actor_1_facebook_likes", y="imdb_score", alpha=0.5, ax=axes[1, 0], color="green")
axes[1, 0].set_title("Likes Pemeran Utama vs IMDb Score")
axes[1, 0].set_xlabel("Actor 1 Facebook Likes")
axes[1, 0].set_ylabel("IMDb Score")

# 4. Negara dengan Jumlah Film Terbanyak (Ungu)
top_countries = df["country"].value_counts().head(10)
sns.barplot(x=top_countries.index, y=top_countries.values, ax=axes[1, 1], palette="Purples_r")
axes[1, 1].set_title("Top 10 Negara dengan Film Terbanyak")
axes[1, 1].set_xlabel("Negara")
axes[1, 1].set_ylabel("Jumlah Film")
axes[1, 1].set_xticklabels(axes[1, 1].get_xticklabels(), rotation=45)

# 5. Jumlah Kritikus vs IMDb Score (Oranye)
sns.scatterplot(data=df, x="num_critic_for_reviews", y="imdb_score", alpha=0.5, ax=axes[2, 0], color="orange")
axes[2, 0].set_title("Jumlah Kritikus vs IMDb Score")
axes[2, 0].set_xlabel("Jumlah Kritikus")
axes[2, 0].set_ylabel("IMDb Score")

# Tampilkan plot
plt.tight_layout()
plt.show()
```
